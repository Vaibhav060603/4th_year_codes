```python
def issafe(board, row, column, n):
    #checking if upper part column is safe or not
    for i in range (row):
        if(board[i][column] == 1):
            return False

    #checking if upper left diagnol is safe or not
    i = row
    j = column
    while i >= 0 and j >= 0:
        if(board[i][j] == 1):
            return False
        i = i-1
        j = j-1

    #checking if upper right diagnol is safe or not
    i = row
    j = column
    while i >= 0 and j < n:
        if(board[i][j] == 1):
            return False
        i = i - 1
        j = j + 1

    return True        
```

```python
def print_solution(board, n):
    for i in range(n):
        print(board[i])
```

```python
def solve(board, row, n):
    if row == n:
        print_solution(board, n)
        print()

    for column in range(n):
        if issafe(board, row, column, n):
            board[row][column] = 1
            solve(board, row+1, n)
            board[row][column] = 0
         
```

```python
n = int(input("Enter the number of queens: "))
board = []

for i in range(n):
    row = []
    for j in range(n):
        row.append(0)
    board.append(row)

solve(board, 0, n)
```
