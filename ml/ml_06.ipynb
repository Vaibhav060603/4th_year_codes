```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Shree\\Downloads\\lp_3_datasets\\sales_data_sample.csv", encoding='latin1')
df
```

```python
df1 = df[["QUANTITYORDERED", "SALES"]]
df1
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
df1 = std.fit_transform(df1)
df1
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range (1,11):
    km = KMeans(n_clusters = i)
    km.fit(df1)
    wcss.append(km.inertia_)     #Inertia: km.inertia_ calculates WCSS, representing the sum of squared distances between each point and its assigned cluster center.
```

```python
from matplotlib import pyplot as plt

plt.plot(range(1,11), wcss)
plt.show()
```

```python
km = KMeans(n_clusters = 3)
km.fit(df1)
plt.scatter(df1[:,0], df1[:,1], c = km.labels_)
```
