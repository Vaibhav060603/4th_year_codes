```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#gradient descent is used of optimization of neural network by minimizing the cost function by 
#finding the minimum error using iterations and learning rate(step size)


# Define the function and its derivative
def f(x):
    return (x + 5) ** 2

def f_derivative(x):
    return 2 * (x + 5)

# Gradient Descent Algorithm
def gradient_descent(starting_point, learning_rate, iterations):
    x = starting_point
    x_values = [x]
    for _ in range(iterations):
        grad = f_derivative(x)          #slope
        x = x - learning_rate * grad    #weight updation for each neuron
        x_values.append(x)
    return x, x_values                  #x -> global minima (contains x from last iterations)
```

```python
# Parameters
starting_point = 10                     
learning_rate = 0.1                     #lesser the learning_rate, more time it take to reach local minima (also called step size)
iterations = 50                         #higher the iterations, more is the chance to reach local minima

# Perform Gradient Descent
min_x, x_values = gradient_descent(starting_point, learning_rate, iterations)
```

```python
# Visualization
x_range = np.linspace(-15, 5, 100)
plt.plot(x_range, f(x_range), color="blue")
plt.scatter(x_values, [f(x) for x in x_values], color="red")

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent on f(x) = (x + 5)^2")
plt.show()
```

```python

#This derivative results in larger gradient values as you move away from the local minima, which can cause the algorithm to overshoot 
#and fail to converge correctly due to the steep slopes.

#with cubic functions like this, it's important to choose a smaller learning rate to prevent oscillations or divergence due to the large gradients.
```
