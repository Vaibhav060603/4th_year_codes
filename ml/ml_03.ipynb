```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Shree\\Downloads\\lp_3_datasets\\bank_Churn_Modelling.csv")
df
```

```python
df = df.drop(columns = ['RowNumber', 'CustomerId', 'Surname'])

from sklearn.preprocessing import LabelEncoder
df["Geography"] = LabelEncoder().fit_transform(df["Geography"])  #LabelEncoder().fit_tranform() coverts the categorical data into unique numbers, like male/female to 0/1
df["Gender"] = LabelEncoder().fit_transform(df["Gender"])
df
```

```python
x = df.drop(columns = ["Exited"])
y = df["Exited"]
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
x = std.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test  = train_test_split(x, y, test_size = 0.3)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#building model
ANN = Sequential([                                                   #to put layers in linear stack fashion
    Dense(16, input_dim = x_train.shape[1], activation = "relu"),    #16, 8, 1 are number of neurons in that layer, input_dim specifies the number of features
    Dense(8, activation = "relu"),                                   #relu -> Rectified Linear Unit (returns input if output is positive, else returns 0)
    Dense(1, activation = "sigmoid")                                 #sigmoid -> (outputs value between 0 and 1)
])

#compiling model
ANN.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ["accuracy"])

#train/fit the model
ANN.fit(x_train, y_train, batch_size = 10, epochs= 15)   #epochs -> means number of pass

#prediction using model
y_pred = ANN.predict(x_test)
x_test
```

```python
y_pred = y_pred>0.5   #convert the y_pred into binary classification into binary (values greater than 0.5 as 1, less than 0.5 as 0)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
ac = accuracy_score(y_test, y_pred)

print("confusion matric: \n", cm, "\naccuracy score: ", ac)
```
